```python
columns_to_remove = ['ip.dst', 'tcp.dstport', 'ip.proto','tcp.flags.syn','tcp.flags.reset','tcp.flags.ack','ip.flags.mf','ip.flags.rb','tcp.seq','tcp.ack','frame.time',]
df = df.drop(columns=columns_to_remove)
```

```python
label_mapping = {
    'Benign': 0,
    'DDoS-PSH-ACK': 1,
    'DDoS-ACK': 1
}

df['Label'] = df['Label'].map(label_mapping)
```

```python
df['Label'].value_counts()
```

```python
from sklearn.utils import resample

majority_class = df[df['Label'] == 1]
minority_class = df[df['Label'] == 0]

downsampled_majority = resample(majority_class, replace=False, n_samples=len(minority_class), random_state=42)
df = pd.concat([downsampled_majority, minority_class])
df = df.sample(frac=1, random_state=42)
df['Label'].value_counts()
```

```python
#PLAN OF ACTION
"""We have handled the missing values, we will split dataset to features and labels, will apply encoding to labels,
will change the ip address coloumn so that it becomes sutiable to apply other mathemetical methods as well as use it as feature
will aplly scaling to feaures(will try with both the available scaling methods which gives better accuracy),
downsampling the majority class to balance the dataset

split into train,test,val

will use pycaret to help us with the best model for us"""
```

```python
X = df[['ip.src', 'tcp.srcport', 'frame.len', 'tcp.flags.push', 'ip.flags.df',
        'Packets', 'Bytes', 'Tx Packets', 'Tx Bytes', 'Rx Packets', 'Rx Bytes']]

Y = df['Label']
```

```python
Y.value_counts()
```

```python
#Lets make an integer representation of the IP address
ip_octets = X['ip.src'].str.split('.', expand=True)

X['ip.src.int'] = (
    ip_octets[0].astype(int) * 256**3 +
    ip_octets[1].astype(int) * 256**2 +
    ip_octets[2].astype(int) * 256**1 +
    ip_octets[3].astype(int)
)
X.head()
```

```python
X_original = X[['ip.src']]

X_to_use= X[['ip.src.int', 'tcp.srcport', 'frame.len', 'tcp.flags.push', 'ip.flags.df',
        'Packets', 'Bytes', 'Tx Packets', 'Tx Bytes', 'Rx Packets', 'Rx Bytes']]
# X_original.head()
X_to_use.head()
```

```python
#Applying both of our decided scaling methods
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

min_max_scaler = MinMaxScaler()
X_min_max_scaled = min_max_scaler.fit_transform(X_to_use)

standard_scaler = StandardScaler()
X_standard_scaled = standard_scaler.fit_transform(X_to_use)

X_min_max_scaled_df = pd.DataFrame(X_min_max_scaled, columns=X_to_use.columns)
X_standard_scaled_df = pd.DataFrame(X_standard_scaled, columns=X_to_use.columns)
```

```python
def int_to_ip_address(ip_int):
    octet_1 = ip_int >> 24 & 255
    octet_2 = ip_int >> 16 & 255
    octet_3 = ip_int >> 8 & 255
    octet_4 = ip_int & 255
    return f"{octet_1}.{octet_2}.{octet_3}.{octet_4}"

ip_int = 3232238849
original_ip = int_to_ip_address(ip_int)
print("Original IP address:", original_ip)
```

```python
#the above function shows how you can get back the IP address which now you can append to list or wherever
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(X_to_use, Y, test_size=0.2, random_state=42)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.4, random_state=42)


dt_classifier = DecisionTreeClassifier(random_state=42)

dt_classifier.fit(X_train, y_train)

val_accuracy = dt_classifier.score(X_val, y_val)

test_accuracy = dt_classifier.score(X_test, y_test)

print(f"Validation Accuracy: {val_accuracy:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
y_pred = dt_classifier.predict(X_test)

report = classification_report(y_test, y_pred)

print("Classification Report:")
print(report)
```

```python
y_pred = dt_classifier.predict(X_test)

predicted_ip_addresses = []

for index, prediction in enumerate(y_pred):
    if prediction == 1:
        ip_int = X_test['ip.src.int'].iloc[index]
        ip_address = int_to_ip_address(ip_int)
        predicted_ip_addresses.append(ip_address)
print("IP addresses predicted as DDOS ATTACKS:")
print(predicted_ip_addresses)
```
